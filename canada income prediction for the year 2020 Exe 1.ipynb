```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
```

```python
df
```

```python
df.head(7)
```

```python
%matplotlib inline
```

```python
plt.xlabel('Year')
plt.ylabel('Per capita income(US$)')
```

```python
plt.xlabel('Year')
plt.ylabel('Per capita income(US$)')
plt.title('Per Capita Income Over Years')
plt.scatter(df['year'], df['per capita income (US$)'])
plt.show()
```

```python
reg = linear_model.LinearRegression() # creating model for linear regression
reg.fit(df[['year']], df[['per capita income (US$)']])
```

```python
reg.predict([[2020]])
```

```python
df
```

```python
reg.coef_ # y = mx + c 

# coef_ is m
```

```python
reg.intercept_ # c
```
